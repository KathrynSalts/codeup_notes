```python
x = list(range(-4,4))
x.append(3.14159)
y = [math.sin(n) for n in x]
# print(x)
# print(y)
plt.scatter(x, y)
plt.plot(x, y, ls=':')

plt.text(0.25, 0, '0, 0', fontsize=10, color='blue')
plt.text(1.2, 0.1, '$\pi$, sin($\pi$))', fontsize=14, color='red')

plt.show()
```

```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c='orange')
plt.plot(x, z, c='green')

plt.annotate('Intersection', xy=(0, 0), xytext=(-3,5), arrowprops={'facecolor':'blue'})
plt.show()
```

```python
x = range(-4, 4)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # width, height

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', ls=':', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', ls=':', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sine and cosine curves')
plt.xlabel('$x$')


plt.show
```

```python
n_rows = 2
n_cols = 2

x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]
z = [1, 2, 3, 4, 5]

#first subplot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y)
plt.title('x ~ y')

#second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x ~ z')

#third subplot
plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y)
plt.title('x ~ y')

#fourth subplot
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, z)
plt.title('x ~ z')
```

```python
x = [1, 1, 2, 3, 3, 3, 4, 4, 5]
bins = [0, 1, 2, 3, 4, 5, 6]
plt.figure(figsize=(16,6))
plt.suptitle(f'x = {x}')

plt.subplot(131)
plt.hist(x, bins=bins)
plt.title('align="middle" (default)')

plt.subplot(132)
plt.hist(x, bins=bins, align='left')
plt.title('align="left" (default)')

plt.subplot(133)
plt.hist(x, bins=bins, align='right')
plt.title('align="right" (default)')

plt.show()
```
